```python
img = cv2.imread( 'sv650.jpg' )
img_grey = img.grey
```

```python
# opencv
import cv2

# initialize feature detector
surf = cv2.SURF(500)

# detect keypoints and descriptions for the image
# img must be uint8
kpoints, desc = surf.detectAndCompute( img, None )
```

```python
redTrans = matplotlib.colors.ColorConverter().to_rgba( (1, 0, 0, .1)  )

plt.imshow( img.mean( axis=2 ) )
plt.gray()
for i in range( 64 ):

    # extract keypoint data
    kp     = kpoints[i]
    width  = kp.size
    height = width
    x = kp.pt[0] - width/2
    y = kp.pt[1] - width/2
    angle  = 0 # being lazy and not visualizing rotation
        
    # make a box
    box = plt.Rectangle( (x,y), width, height, 
                         angle=angle, ec='red', 
                         fc=redTrans, linewidth=2 )

    # add it to the plot
    plt.gca().add_patch(box)
```

```python
plt.rcParams['figure.figsize'] = (16.0, 16.0)

for i in range(64):
    kp = kpoints[i]
    xmin = max( round( kp.pt[0] - kp.size/2. ), 0 )
    ymin = max( round( kp.pt[1] - kp.size/2. ), 0 )
    xmax = min( round( kp.pt[0] + kp.size/2. ), img.shape[1]-1 )
    ymax = min( round( kp.pt[1] + kp.size/2. ), img.shape[0]-1 )
    
    patch = img[ymin:ymax,xmin:xmax,(2,1,0)]
    plt.subplot( 8, 8, i+1 )
    plt.imshow( patch )

plt.rcParams['figure.figsize'] = (16.0, 12.0)
```
