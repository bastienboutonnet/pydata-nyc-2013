```python
import scipy as sp
from sklearn import cluster
from sklearn import metrics

MAX_DIM = 512

def extract_feature( img_filename, feature_extractor ):
#    print img_filename   
    # extract local feature (SURF) from one image
    img = cv2.imread( img_filename , cv2.CV_LOAD_IMAGE_GRAYSCALE)
    
    # resize if too big
    if max( img.shape ) > MAX_DIM:
        scale = MAX_DIM / float( max( img.shape ) )
        img = sp.misc.imresize( img, scale )
    
    # return keypoints and descriptor
    kpoints, desc = feature_extractor.detectAndCompute(img,None)
    return kpoints, desc
```

```python
# extract all local features from a number of files
def create_flist(file_list , feature_extractor): 
    
    desc_list = []
    for f in file_list:
        kpoints, desc = extract_feature(f, feature_extractor)
        desc_list.append(desc)

    # array of features from all file
    flist = sp.vstack( tuple(desc_list) ).astype( 'double' )
    return flist
    
```

```python
class VisualVocabulary(object):
    
    def __init__( self, vsize=0 ):
        self.vsize = vsize

    def fit( self, flist ):
        # create a k-means of vsize clusters
#        self.centroids, labels, inertia = cluster.k_means( flist, self.vsize )
        mbkm = cluster.MiniBatchKMeans( self.vsize )
        # fit to the set of provided feature descriptors
        mbkm.fit( flist )
        # retain the centroids the centroids
        self.centroids = mbkm.cluster_centers_

    def quantize( self , flist ):
        # get the distance to all of the cluster centroids
        dist = metrics.pairwise_distances( self.centroids, flist )
        # quantize to the closest cluster
        nearest_centroids = sp.argmin( dist , axis=0 )

        # generate histogram given that
        hist, dummy = sp.histogram( nearest_centroids, self.vsize, [0, self.vsize] )
        return hist
```

```python
class BOW(object):

    def __init__(self, file_list, feature_extractor, vsize ):
        # keep stuff
        self.feature_extractor = feature_extractor
        self.vsize = vsize

        # create the array of feature descriptors from the filenames
        flist = create_flist( file_list , self.feature_extractor)
        # init the vocabulary
        self.vv = VisualVocabulary( self.vsize )
        # fit to the provided feature descriptors
        self.vv.fit( flist )

    def represent( self, img_filename ):
        # extract local features
        kpoints, desc = extract_feature(img_filename, self.feature_extractor)

        # represent the object image as histogram of visual vocabulary
        obj_repr = self.vv.quantize(desc)
        return obj_repr
```

```python
surf = cv2.SURF(500)
```

# Object classification

```python
motoFiles = glob.glob( 'photos/Motorbikes/*' )
pianoFiles = glob.glob( 'photos/grand_piano/*' )
```

```python
%time bowMoto = BOW( motoFiles[:150] + pianoFiles[:150], surf, 400 )
```

```python
featureFiles = motoFiles[:300] + pianoFiles
```

```python
X = bowMoto.represent( featureFiles[0] )
X = np.zeros( ( len( featureFiles ), X.shape[0] ) )

for i,f in enumerate(featureFiles):
   X[i] = bowMoto.represent( f )
```

```python
y = []
y += [ 1 for i in range(300) ]
y += [ 0 for i in range(len(pianoFiles)) ]
y = np.array( y )

print X.shape
print y.shape
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test, f_train, f_test = train_test_split( X, y, featureFiles )

print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

```python
from sklearn import svm
from sklearn.cross_validation import cross_val_score
print cross_val_score( svm.SVC(), X, y, cv=5, scoring='precision')
```

```python
clf = svm.SVC()
clf.fit(X_train, y_train)
results = clf.predict( X_test )
```

```python
plt.rcParams['figure.figsize'] = (16.0, 16.0)

for i in range(64):
    plt.subplot( 8, 8, i+1 )
    tmpImg = sp.misc.imread( f_test[i] )
    
    plt.imshow( tmpImg )
    frame1 = plt.gca()
    frame1.axes.get_xaxis().set_ticks([])
    frame1.axes.get_yaxis().set_ticks([])

    color = 'black' if (results == y_test)[i] else 'red'
    text  = 'Moto' if results[i] else 'Piano'
    plt.title( text, fontsize='small', color=color )

#     xlabel( str( results[i] ) )
    
plt.rcParams['figure.figsize'] = (16.0, 12.0)
```
