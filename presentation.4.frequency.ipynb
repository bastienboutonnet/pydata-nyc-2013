```python
import numpy as np
import scipy as sp
import scipy.ndimage
import scipy.fftpack
import skimage.io
import skimage.exposure
import skimage.color
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16.0, 8.0)
```

## Frequency space analysis

```python
img2 = skimage.io.imread( 'sv650.jpg', as_grey=True )
```

```python
imgfft = img2.astype( float )
```

```python
fftres = sp.fftpack.fft2( imgfft )
```

```python
fftres[:2,:2]
```

```python
plt.imshow( np.log10( 1 + abs( sp.fftpack.fftshift( fftres ) ) ) )
```

```python
imgSCV = skimage.io.imread( 'mono_road.jpg', as_grey=True )

imgfft2 = imgSCV.astype( float )
fftres2 = sp.fftpack.fft2( imgfft2 )
```

```python
plt.imshow( np.log10( 1 + abs( sp.fftpack.fftshift( fftres2 ) ) ) )
```

```python
noise = sp.random.uniform( size=(512,512) )
nfft = sp.fftpack.fft2( noise )
```

```python
frSm = sp.ndimage.gaussian_filter1d( abs(fftres[0,:fftres.shape[1]/2]), 5 )
frSm2 = sp.ndimage.gaussian_filter1d( abs(fftres2[0,:fftres2.shape[1]/2]), 5 )
frSmN = sp.ndimage.gaussian_filter1d( abs(nfft[0,:nfft.shape[1]/2]), 5 )
```

```python
def shift_blur( arr ):
    arr2 = abs( arr )
    arr2 = sp.fftpack.fftshift( arr2 )
    arr2 = sp.ndimage.gaussian_filter1d( arr2, 5 )
    arr2 = sp.fftpack.fftshift( arr2 )
    arr2 = arr2[ :arr2.shape[0] / 2 ]
    return arr2
```

```python
arr = fftres[0]
arr2 = abs( arr )
arr2 = sp.fftpack.fftshift( arr2 )
```

```python
plt.semilogy( arr2 )
```

```python
frSm = shift_blur( fftres[0] )
frSm2 = shift_blur( fftres2[0] )
frSmN = shift_blur( nfft[0] )
```

```python
plt.hold( True )
plt.semilogy( np.linspace( 0, 1, frSm.shape[0] ), frSm / frSm.mean() )
plt.semilogy( np.linspace( 0, 1, frSm2.shape[0] ), frSm2 / frSm2.mean() )
plt.semilogy( np.linspace( 0, 1, frSmN.shape[0] ), frSmN / frSmN.mean() )
```
