```python
import numpy as np
import scipy as sp
import skimage.io
import skimage.exposure
import skimage.color
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16.0, 8.0)
```

# Gradients and Edges

```python
flower = skimage.io.imread( 'macro-flower.jpg' )
```

```python
sp.gradient?
```

```python
flower_g = sp.diff( flower.mean(axis=2) )
```

```python
plt.imshow( flower_g )
plt.gray()
```

```python
flower_g = sp.gradient( flower.mean(axis=2) )
```

```python
plt.subplot( 1, 2, 1 )
plt.imshow( flower_g[0] )
plt.subplot( 1, 2, 2 )
plt.imshow( flower_g[1] )
plt.gray()
```

```python
plt.imshow( np.sqrt( flower_g[0]**2 + flower_g[1]**2 ) )
```

```python
plt.imshow( np.log( 1 + np.sqrt( flower_g[0]**2 + flower_g[1]**2 ) ) )
```

```python
highway_g = sp.gradient( skimage.io.imread( 'mono_road.jpg' ).mean( 2 ) )
plt.subplot( 1, 2, 1 )
plt.imshow( np.sqrt( highway_g[0]**2 + highway_g[1]**2 ) )
plt.subplot( 1, 2, 2 )
plt.imshow( np.sqrt( highway_g[0]**2 + highway_g[1]**2 ).clip( 0, 25 ) )
```
