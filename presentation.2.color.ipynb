# Color statistics

## RGB histograms

```python
img = skimage.io.imread( 'mono_road.jpg' )
img = img.astype( float ) / 255
```

```python
plt.imshow( img )
```

```python
hist = skimage.exposure.histogram( img.mean(axis=2) )
plt.plot( hist[1], hist[0] )
```

```python
hist_rgb = [ skimage.exposure.histogram( img[...,i] ) for i in range(3) ]
```

```python
RGB_COLORS = [ 'r', 'g', 'b' ]
plt.hold( True )
for i in range(3):
    plt.plot( hist_rgb[i][1], hist_rgb[i][0], RGB_COLORS[i] )
```

## Hue-only image

```python
# convert to HSV space
hue_only = skimage.color.rgb2hsv( img )

# set saturation and value to uniform .5
hue_only[...,1] = .5
hue_only[...,2] = .5

# convert back
hue_only_rgb = skimage.color.hsv2rgb( hue_only )

# plot
plt.imshow( hue_only_rgb )
```

## Image color space visualization

```python
rgbTriples = []
imgNp = imgSCV.getNumpy()
for i in range( 0, imgNp.shape[0], 20 ):
    for j in range( 0, imgNp.shape[1], 20 ):
        rgbTriples += [ imgNp[i,j,...].astype(float) / 255. ]
        
Xs, Ys, Zs = zip( *rgbTriples )
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter( Xs, Ys, Zs, s=40, c=rgbTriples )
```

```python
labTriples = []
labNp = skimage.color.rgb2lab( imgNp )
for i in range( 0, labNp.shape[0], 20 ):
    for j in range( 0, labNp.shape[1], 20 ):
        labTriples += [ labNp[i,j,...] ]

Xs, Ys, Zs = zip( *labTriples )
```

```python
flower = skimage.io.imread( 'macro-flower.jpg' )
plt.imshow( flower )
```

```python
labflower = skimage.color.rgb2lab( flower )

rgb_points = []
lab_points = []

for i in range( 0, flower.shape[0], 20 ):
    for j in range( 0, flower.shape[1], 20 ):
        rgb_points += [ flower[i,j,...].astype(float) / 255. ]
        lab_points += [ labflower[i,j,...] ]

Xsm, Ysm, Zsm = zip( *rgb_points )
Xsl, Ysl, Zsl = zip( *lab_points )
```

```python
plt.rcParams['figure.figsize'] = (10.0, 10.0)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter( Xsm, Ysm, Zsm, s=20, c=rgb_points )
plt.rcParams['figure.figsize'] = (16.0, 8.0)
```
