```python
# import opencv 2 python bindings
import cv2

# load and image and convert to gray
group = skimage.io.imread( 'group-shot.jpg' )
group_grey = group.mean( axis=2 ).astype( 'uint8' )
```

```python
plt.imshow( group )
```

```python
# inital Haar wavelet cascade classifier with frontal face data
cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# detect faces
rects = cascade.detectMultiScale(group_grey, 1.3, 5 )

# number of faces
print len( rects )
# print first face data
rects[0]
```

```python
for x1, y1, x2, y2 in rects:
    cv2.rectangle( group, (x1, y1), (x1+x2, y1+y2), (127, 255, 0), 2)
```

```python
plt.imshow( np.asarray( group ) )
```

```python
# load profile face Haar cascadeq`
cascade = cv2.CascadeClassifier('haarcascade_profileface.xml')
rects = cascade.detectMultiScale(group_grey, 1.3, 5 )
```

```python
for x1, y1, x2, y2 in rects:
    cv2.rectangle( group, (x1, y1), (x1+x2, y1+y2), (192, 0, 0), 2)
    
plt.imshow( np.asarray( group ) )
```
