# First-order statistics

# Pixel values

```python
# scikits image
img = skimage.io.imread( 'sv650.jpg', as_grey=True )
img_lo = skimage.io.imread( 'sv650_under.jpg', as_grey=True )
img_hi = skimage.io.imread( 'sv650_over.jpg', as_grey=True )
```

```python
plt.gray()
plt.subplot( 1, 3, 1 )
plt.imshow( img_lo )
plt.subplot( 1, 3, 2 )
plt.imshow( img )
plt.subplot( 1, 3, 3 )
plt.imshow( img_hi )
```

```python
print img.mean()
print img_lo.mean()
print img_hi.mean()
```

```python
img.max()
```

```python
print np.count_nonzero( img < .05 ) / img.size * 100
print np.count_nonzero( img_lo < .05 ) / img_lo.size * 100
```

```python
plt.subplot( 1, 2, 1 )
plt.imshow( ~(img < .05 ) )
plt.subplot( 1, 2, 2 )
plt.imshow( ~(img_lo < .05) )
```

```python
plt.subplot( 1, 2, 1 )
plt.imshow( img > .98 )
plt.subplot( 1, 2, 2 )
plt.imshow( img_hi > .98 )
```

```python
print np.count_nonzero( img > .98 ) / img.size * 100
print np.count_nonzero( img_hi > .98 ) / img_hi.size *100
```

## Histograms

- __NOTE__: ``histogram`` operates on a flattened image
- evaluates all color channels

```python
hist = skimage.exposure.histogram( img, nbins=256 )
print hist[0][:5]   # count per bin
print hist[1][:5]   # center of bin
```

```python
plt.plot( hist[1], hist[0] )
plt.axis( [0, 1, 0, 18000] )
```

```python
hist_lo = skimage.exposure.histogram( img_lo )
hist_hi = skimage.exposure.histogram( img_hi )
```

```python
plt.hold( True )
plt.plot( hist[1], hist[0] )
plt.plot( hist_lo[1], hist_lo[0], ':' )
plt.plot( hist_hi[1], hist_hi[0], '--' )
plt.legend( ['original', 'under-exposed', 'over-exposed'] )
```
